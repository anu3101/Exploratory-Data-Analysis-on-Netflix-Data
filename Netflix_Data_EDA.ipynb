```python
#AVERAGE DURATION BY RATINGS
df['Duration_min'] = df['Duration'].apply(lambda x: int(x.split()[0]) if 'min' in x else None)
df.groupby('Rating')['Duration_min'].mean().dropna()
```

```python
#Longest movie and Shortest Movie in Data
longest_movie = df.loc[df['Duration_min'].idxmax()]
shortest_movie = df.loc[df['Duration_min'].idxmin()]
```

```python
sns.countplot(x='Rating', data=df, color= "#615EFC", order=df['Rating'].value_counts().index)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#TOP 10 CATEGORIES
top_categories = df['Category'].value_counts().head(10)
sns.barplot(x=top_categories.index, y=top_categories.values)
plt.title('Top 10 Most Common Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
releases_by_year = df['Year'].value_counts().sort_index()
sns.lineplot(x=releases_by_year.index, y=releases_by_year.values, marker='o')
plt.title('Number of Releases Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.show()
```
